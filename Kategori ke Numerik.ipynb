## Encode ke numerik (kolom yang dipilih)

Dimulai dari label yang ter-encode

Label encoder

```
setColumn1 = columnSelected["path"]
label_encoder_setColumn1 = preprocessing.LabelEncoder()
numeric_setColumn1 = label_encoder_setColumn1.fit_transform(setColumn1)

print(numeric_setColumn1)
```

1 hot encoder (satu enkoder panas)

```
one_hot_encoder = pd.get_dummies(setColumn1)

print(one_hot_encoder)
```

Ordinal Encoder (Enkoder Tingkatan; tidak terlalu diterapkan pada data yang digunakan)

```
numeric_setColumn1_1 = [numeric_setColumn1]
ordinal_encoder = preprocessing.OrdinalEncoder()
numeric_setColumn1_1_ordinal = ordinal_encoder.fit_transform(numeric_setColumn1_1)

print(numeric_setColumn1_1_ordinal)
```

Targetting Encoder (Enkoder Terjang)

```
data2 = pd.DataFrame({"character" : columnSelected["path"], "Target" : columnSelected["rarity"]})

mean_target_data2 = data2.groupby("character")["Target"].mean()
data2["Target"] = data2["character"].map(mean_target_data2)

print(data2)
```
